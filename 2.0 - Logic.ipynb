Program flow is the art of controlling which parts of your code get executed when, in what order, how many times, under what conditions, where to start and stop, etc.

This is essential to making sure your program actually does what you want it to!

Flow is controlled mainly by using **conditional logic**.

One of the distinctive features of Python is that conditions are created mainly using **indentation**.

This makes Python code very easy to read, as you will see.

Standard practice is that each indent is 4 characters long. You can press space 4 times, but in many editors (including Jupyter) the tab key will give you the 4 spaces in one go.

## Conditional statements

### The **if** statement

An `if` statement is a block of code that first checks if a specified condition is true, and only in that case will it carry out a task.

Let's look at an example.

```python
# Variable specifying a type of animal
animal = 'tiger'


# AN IF STATEMENT:

if animal == 'tiger': # The first line specifies the condition to test
    
    # If the condition is true, execute the (indented) code below
    print('Run!')

    
# The code below here is no longer indented, so the if statement is finished
# Therefore, this code will always run:
print('\nDone.')

```

You can try changing the string contained in `animal` to test the behaviour of the `if` statement.

---

All `if` statements:
  - start with 'if'
  - use a colon after the condition
  - **apply only to the indented code**
  
Note also the double equals sign, `==`, in the condition. This is always used in conditions that test for equality, to differentiate from the single equals sign that would try to assign a new value!

Other conditional operators include:
  - `>=` : greater than or equal to
  - `<=` : less than or equal to
  - `!=` : not equal to
  - `in` : check if value is in a given list or string
  
Conditions always evaluate to `True` or `False`, e.g.

```python
result = (animal=='tiger')

print(result)
```

### **If-else**

If statements only execute if a condition is true.

Sometimes, we want to execute a block of code only if a condition is false. For this, we can use an if-else.

The syntax starts off the same, but has an extra `else:` statement:

```python
# Variable specifying a type of animal
animal = 'dog'

danger_list = ['tiger', 'lion', 'shark']


# Condition:
if animal in danger_list:
    # If the condition is true, execute the (indented) code below
    print('Run!')
else:
    # If the condition is false, execute this code instead:
    print('Pet!')

    
print('\nDone.')
```

---

### **Elif**

A natural extension of this is `if`-`elif`-`else`, where elif is short for 'else if'

The middle block now tests another condition before executing:

```python
# Variable specifying a type of animal
animal = 'frog'


danger_list = ['tiger', 'lion', 'shark']

safe_list = ['dog', 'cat', 'rabbit']


# Condition:
if animal in danger_list:
    # If the condition is true, execute the (indented) code below
    print('Run!')
elif animal in safe_list:
    # If the first condition is false, but the second condition is true, execute this code instead:
    print('Pet!')
else:
    print('Unknown animal! Observe carefully')
    

print('\nDone.')
```

```python
# A simple for loop:

# A list of the Soprano family
sopranos = ['Tony', 'Carmella', 'Meadow', 'AJ']

# Iterate over this list
for person in sopranos:
    # Execute this code for each person
    # the value of person is different each time
    print(person)
```

### Using loops for incrementing

Any variable in the loop will be updated each time it runs. This has many handy applications, e.g. keeping track of how many times the loop has executed.

Note that in the example below, we use the common shorthand:

    y += x
    
Which is equivalent to

    y = y + x
    
i.e., the new value of `y` is just the previous value plus `x`

```python
# Create a variable to act as a counter, and initialize it to zero
family_size = 0

# Add one to the counter each time we go through the loop:
for person in sopranos:
    family_size += 1
        
print('Number of family members =', family_size)
```

---

### Mathematical operations with a list

For loops also provide a flexible way to do maths with the list elements themselves.

In this example, we will sum an arbitrary list. Note that we could also do this with `list.sum()`, but in a loop we can adapt the code to do other things too!

```python
my_list = [0.33, 0.67, 2.33, 6.67]

# Create a counter
s = 0

for i in my_list:
    # Add the current value of i to the counter each time
    s += i
    
    # print the current value of the counter
    # This time, we indent the print statement so that it shows a running count
    print(s)
```

If we want to iterate a given number of times, we don't always need to create and store a list.

It is easier and more common to use the `range()` function.

`range(x)` will provide a list of integers of length `x`, starting with `0`.

```python
# Example of the range command
for i in range(10):
    print(i)
```
