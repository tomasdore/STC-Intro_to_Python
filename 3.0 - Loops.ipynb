## Repeated statements: loops

### **for** loops

Now we will look at loops. A loop is a block of code that will iterate (execute consecutively) multiple times.

A `for` loop is the simplest loop. It requires something to iterate over, i.e. an "iterable" like a list (do something for every item in the list) or a string (do something for every character in the string


```python
# A simple for loop:

# A list of the Soprano family
sopranos = ['Tony', 'Carmella', 'Meadow', 'AJ']

# Iterate over this list
for person in sopranos:
    # Execute this code for each person
    # the value of person is different each time
    print(person)
```

---

### Using loops for incrementing

Any variable in the loop will be updated each time it runs. This has many handy applications, e.g. keeping track of how many times the loop has executed.

Note that in the example below, we use the common shorthand:

    y += x
    
Which is equivalent to

    y = y + x
    
i.e., the new value of `y` is just the previous value plus `x`

```python
# Create a variable to act as a counter, and initialize it to zero
family_size = 0

# Add one to the counter each time we go through the loop:
for person in sopranos:
    family_size += 1
        
print('Number of family members =', family_size)
```

---

### Mathematical operations with a list

For loops also provide a flexible way to do maths with the list elements themselves.

In this example, we will sum an arbitrary list. Note that we could also do this with `list.sum()`, but in a loop we can adapt the code to do other things too!

```python
my_list = [0.33, 0.67, 2.33, 6.67]

# Create a counter
s = 0

for i in my_list:
    # Add the current value of i to the counter each time
    s += i
    
    # print the current value of the counter
    # This time, we indent the print statement so that it shows a running count
    print(s)
```

### What if we don't have a list handy?

If we want to iterate a given number of times, we don't always need to create and store a list.

It is easier and more common to use the `range()` function.

`range(x)` will provide a list of integers of length `x`, starting with `0`.

```python
# Example of the range command
for i in range(10):
    print(i)
```

## Nested loops/statements

We can 'nest' loops and conditional statements inside each other for fine control of the program

E.g. we can put one `for` loop inside another:


```python
# Outer loop executes for each of the Sopranos
for person in sopranos:
    # Each time the outer loop executes, the inner loop here executes for each letter of their name
    for letter in person:
        print(letter)
```

## Special words for loops

The following words can also be used to control the flow of a loop:

  - `pass` : do nothing
  - `continue` : stop this iteration of the loop early, and go on to the next one
  - `break` : end the loop entirely
  
The following examples show how to use these words:

```python
# Pass:

for i in all_animals:
    if i in safe_list:
        # Do nothing if the animal is safe
        pass
    else:
        print('Danger:', i)
```

```python
# Continue:

for i in all_animals:
    print(i)
    if i in safe_list:
        # If the animal is safe, skip to the next iteration
        continue
    # The rest of the loop will be skipped on iterations that reach the continue
    print(' - danger!')
    
```

```python
# Break:

for i in all_animals:
    print(i)
    if i in danger_list:
        print(' - too scary! Stopping loop and running away')
        break
```

## **while** loops

We can also create a loop that will continue indefinitely, as long as a certain condition is true. This is often more convenient than creating nested `for` and `if` loops.

The `while` loop can be seen in the example below:

```python
my_counter = 0

while my_counter <= 5:
    print(my_counter)
    my_counter += 1
```

We have to be very careful with `while` loops. If we were to remove the increment `my_counter += 1` from the above example, the loop would keep printing `0` forever (and slow down our code to a halt).

---

## Summary

- Conditional logic means statements that evaluate to `True` or `False`
- An `if` statement will only execute if a condition is `True`
- A `for` loop repeats the specified code a set number of times. This could be e.g. repeating for every item in a list
- A `while` loop repeats indefinitely as long as a condiation remains `True`
