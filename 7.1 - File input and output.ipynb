We often wish to use code for operations on files, for example reading an input file containing data or parameters, or saving results to a text file for access later. In this lecture we will learn how to read and write text files in Python.


## Opening a file

The first step is usually to open the file. We do this using the `open()` function. 

This function allows us to specify different modes depending on how we plan to interact with the file:


- 'r' - opens a file for reading only.
- 'w' - opens a file for writing. If the file exists, it overwrites it. Otherwise, it creates a new file.
- 'a' - opens a file for appending only. If the file doesn't exist, it creates the file.
- 'x' - creates a new file. If the file exists, it fails.
- '+' - opens a file for updating.

When we are finished with the file, we have to close it. This releases memory used in opening the file. 

When writing to a file, changes are not saved until the file is closed!

We close the file using the method `.close()` on open files.

## Reading from a text file

Let's first examine how to read from a file.

The basic way:

```python
# Open the file called zen_of_python.txt in READ mode, call the open file object 'f':

f = open('zen_of_python.txt', 'r')
```

We can now look at the contents of the file with two methods on the file object:

`.read()` will read the entire contents of the file

`.readline()` will read only the next line, and can be called repeatedly until the entire file has been read

```python
# Print the entire file:
print(f.read())
```

---

Try executing the above cell a second time. The output will be blank. **Once the file object has been read to the end, any subsequent calls return an empty string**.

```python
# Now close the file!

f.close()
```

Now let's try printing one line at a time

```python
# We need to open the file again:

f = open('zen_of_python.txt', 'r')
```

```python
# Print one line at a time (execute this cell repeatedly):

print(f.readline())
```

Each time we run the above cell, it prints a new line.


```python
# Remember to close the file

f.close()
```

---

We can also read all lines to a list using the `.readlines()` method:

```python
f = open('zen_of_python.txt', 'r')

lines = f.readlines()

f.close()

print(lines)
```

## The safer way to open files

We can make sure that files are only open for as long as we need them by using a `with` statement. This is the preferred syntax (shorter as well as safer!)

```python
# Use 'with'...'as' to wrap around file operations
with open('zen_of_python.txt', 'r') as f:
    # file operations go here
    print(f.read())
```

## String formatting

To write numeric data (for example) to a text file, we use string formatting, e.g.:


```python
euler = 2.718281828459

with open('my_file.txt', 'w') as f:
    f.write('%f' %euler)
```

The first `%` sign (inside the string) indicates that we are writing a variable. The letter that follows indicates what type of variable:

- %s = string
- %d = integer
- %f = float
- %e = float, but use scientific notation

We can also specify how many decimal places to include, e.g. %.2f will write a float to 2 decimal places!

The second % sign (after the string) tells your code which variable to write at the first % sign

For example:

```python
euler = 2.718281828459

with open('my_file.txt', 'w') as f:
    f.write('%.2f' %euler)
```

You can include as many variables as you want by putting several % signs in your string, and providing a tuple after the string:


```python
length = 6
datatype = 'string'

with open('my_file.txt', 'w') as f:
    f.write('This is a %d word %s' %(length, datatype))
```

You should also be aware of two other special characters:

- `\t` represents a tab
- `\n` represents a new line

All of these formatting techniques also work with `print()` !

```python
length = 6
datatype = 'string'

print('This is a %d word %s \nAnd this is a new line' %(length, datatype))
```

## Reading specialised data types

Sometimes you will want to read in data that is not a simple text file, e.g.

- A CSV (comma-separated variables) file: this is like a table or spreadsheet, with columns separated by commas
- A JSON file: this is structured like a Python dictionary

It is best to read these using specialised modules. We will learn two ways to read CSV files later in the course.

To read and write JSON, we use the json module:

```python
import json

masses = {'Earth': 1.0 , 'Jupiter': 317.8, 'Mercury': 0.0553, 'Mars': 0.107, 'Venus': 0.815}

# Use json.dump to write to file
with open('planets.json', 'w') as f:
    json.dump(masses,f)
```

Let's look at the file we created.

Notice that all the quotation marks have become double quotes -- this is the standard used by JSON

```python
# Use json.load to read from file
with open('planets.json', 'r') as f:
    new_masses = json.load(f)

# Check that we successfully read the JSON dictionary:
print(new_masses)
```

---

### Reading a dictionary from a string with JSON

Sometimes you will find dictionaries stored as if they were strings, e.g. if the dictionary is one entry within a larger database.

To turn a dictionary into a string (like with any other data type) we simply add quotation marks:

```python
# Create a string that looks like a dictionary:
dict_string = '{"Earth": 1.0 , "Jupiter": 317.8, "Mercury": 0.0553, "Mars": 0.107, "Venus": 0.815}'

# Check its type:
print(type(dict_string))
```

*Notice the use of both double and single quotation marks here: if we start a string with a single quote, we can use double quotes inside the string, or vice versa. But if we try to use the same type of quote both around and within the string, it would end the string early!*

We can turn this into a dictionary with `json.loads`, where the extra 's' is for string:

```python
# Use json.loads to load a dictionary from a string
dict = json.loads(dict_string)

# Check its type:
print(type(dict))
```
