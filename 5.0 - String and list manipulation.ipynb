## Indexing in Python

Picking out one value from a series can usually be done by giving your code the **index** of the entry you want. The index just says at which position in the series your entry lies.

**Very important: in Python, the first entry in an ordered sequence has index zero, the next has index one, etc.** 

This is called zero-based indexing, and differs from some other languages where indices start at one instead.

We can also pick out items starting from the end and working backwards, using negative indices. The last item is -1, the second-last is -2, etc.

## String slicing

Let's start with the simple example of "string slicing". This means extracting certain characters from a string, but the logic is similar to working with lists.

```python
# Define a string to work with
my_string = 'A string of five words'

print(my_string)
```

Let's look at the indices of characters in the string:

    A string o...
    0123456789...


### Accessing a single character

We access specific parts of the string using the index with a square bracket notation:


```python
# Print the first character in the example string
print(my_string[0])
```

### String 'immutability'

Although we can access parts of a string to see their value, we cannot change them in place.

e.g. the following will cause an error:

```python
# This is not allowed:

# my_string[0] = 'B'
```

---

### Accessing a sequence (sub-string) of any length

We can access a sequence of characters by specifying a range to slice. Ranges use a : notation, e.g. 

    [1:10]

Things to note:
  - The 'slice' occurs **before** each index
    - So e.g. `[1:10]` would cut the string between 0 and 1, and again between 9 and 10, returning characters 1-9 
  - Spaces count as characters

---

Correct the indices below to print the first two words in our example ("A string"):

```python
lower_index = 1

upper_index = 10

print(my_string[lower_index:upper_index])
```

Solution: We have to slice before character 0, and before character 8, hence our slice is

    [0:8]
    
---

### One-sided cuts

The above example returns a section between two indices. We can also return all characters before or after a particular index by leaving one side of the slice blank:


```python
# Print only characters at positions before 8:

print(my_string[:8])
```

```python
# Print only characters at positions 8 or higher:

print(my_string[8:])
```

---

### Creating a new string with slicing

And of course, we can store our sub-string as a new variable (and manipulate that in the same way if desired):

```python
# Define a new variable representing the sub-string:
sub_string = my_string[:8]
print(sub_string)

# Slice the new sub-string, e.g.
print(sub_string[2:])
```

### String splitting

A very useful method for manipulating strings is splitting, i.e. breaking a string into multiple parts.

The syntax is best understood with an example:

```python
# A single string:
print(my_string)

# The same string broken up at a specified character, in this case at spaces:

split_string_1 = my_string.split(' ')

print(split_string_1)
```

```python
# We can instead choose to split at a different character, e.g.:

split_string_2 = my_string.split('g')

print(split_string_2)

# Try changing the character to split around!
```

## Tuples

A 'tuple' (think of 'multiple') is a Type that holds an arbitrary sequence of items, which can be of different types.

They are specified using rounded brackets, with each item separated by a comma

```python
# This is a valid tuple
my_tuple = ('A', 'tuple', 'of', 5, 'entries')
# Note that it contains both strings and an integer
```

The items in a tuple can be accessed using similar notation as the characters in a string

```python
print(my_tuple[0])
```

---

## Lists

While tuples use only a small amount of memory, they have one drawback: once created, the items cannot be changed -- tuples are immutable, like strings.

A similar, but more flexible, data Type is a **list**

These are also comma-separated, but use square brackets.

You have already seen one of these: the string splitting method above returned a list!

```python
# Define a list
my_list = ['A', 'list', 'of', 5, 'entries']

print(my_list)
```

Lists support **assignment**: you can access an item and change its value:

```python
# Access and replace individual entries in the list with new values:
my_list[2] = 'with'
my_list[3] = 4
my_list[4] = 'strings'

print(my_list)
```

---
Items in a list can also be lists in themselves:

```python
my_list[3] = 'a'
my_list[4] = ['nested', 'list']

print(my_list)
```

---

To access values from a list within a list, you just repeat the square bracket notation.

Say we want the first item in the inner list. We can return the inner list with its index `[4]`, and then the first item from there with its index, `[0]`:

```python
# Step-by-step approach
inner_list = my_list[4]

print(inner_list)

print(inner_list[0])
```

```python
# All-in-one approach -- just add another square bracket!
print(my_list[4][0])
```

---
If you try to access an index that does not exist, you will get an error.

E.g. our list has 5 items, so the only valid indices are `0-4`:

```python
#  This will not work:

# my_list[6]
```

### Convenient functions for lists

Python includes several handy functions for working with lists. E.g. to return the length of a list:

```python
new_list = [1, 1, 2, 3, 5, 8, 13, 21]

print(len(new_list))
```

For numerical lists, we can easily compute the sum:

```python
print(sum(new_list))
```

Other example functions:

```python
print(min(new_list))
print(max(new_list))
```

---

Lists also contain a number of handy **methods**. We will learn more about methods later, but essentially these are functions built-in to a class of objects.

We will not go through all the methods, but here are two examples.

1. Make a copy of a list:

```python
copied_list = new_list.copy()

print(new_list)
print(copied_list)
```

---

2. Append a new entry to a list:


```python
copied_list.append(34)

print(copied_list)
```

You can find many more handy methods to remove, insert and rearrange list items in the Python documentation [here](https://docs.python.org/3/tutorial/datastructures.html). 



```python
# Initialise a dictionary, with some keyword-value pairs

planet_masses = {'Earth': 1.0 , 'Jupiter': 317.8, 'Mercury': 0.0553}
# These are the masses of some planets, relative to the Earth

print(planet_masses)
```

We can access any value using the keyword, similar to how we used indices for lists. 

The advantage of a dictionary is that we don't need to care about *where* the value we want is, we just have to remember what we called it!

```python
# Remember that the name of the key is always a string, and needs quotation marks!
print(planet_masses['Jupiter'])
```

We can easily add new entries to a dictionary:

```python
planet_masses['Mars'] = 0.107

print(planet_masses)
```

Dictionaries are very useful for storing and retrieving data, but they do not permit e.g. easy math operations like lists
