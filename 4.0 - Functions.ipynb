## Functions

Very often in coding, we want to do the same operation at different times or with different data. You could do this by copy-pasting your code. But it is much better to do this by defining re-usable **functions**.

Functions are separate, named blocks of code for a specific purpose. The code inside a function is 'walled off' from your main code.

Your main code can call the function (as many times as you want) simply by using the name of the function and supplying any inputs it requires.

You have already seen many built-in Python functions, e.g. `print()`!

Every function has a name, a list of (required and optional) inputs in parentheses, and returns something at the end.

---

To create a new function, we use the syntax:

```python
# def means define

def my_function(): # any input parameters go in the brackets after the name of the function
    
    # The function code to execute goes here
        
    return # this ends the function, and returns any object specified to the main code

```

**Just like in loops, Python uses indentation to know what is part of the function.**

We will build up our first real function from here.

---

A function can have as many (or as few) input parameters as we want. These inputs are called 'keyword arguments'.

Let's imagine some simple examples:

```python
# A function with no argments:

def hello_world():
    print('Hello, world!')
    return
```

We call a function using its name including the brackets -- just like `print()`!

```python
hello_world()
```

---

### Functions with a single argument

Providing an argument allows the user to specify the behaviour when they call the function. An obvious example is providing a string for the `print()` function.

Let's make our own example:

```python
# A function with one argument:

def repeat_world(n):
    i = 0
    while i < n:
        print('Hello, world!')
        i += 1
    return
        
```

The user provides a value for `n` when they call the function, in this case specifying how many times they want to print the message

```python
repeat_world(n=5)
```

---

What happens if we don't provide an argument?

```python
repeat_world()
```

This function requires an argument, so we get an error message!

---

### Functions with multiple arguments

Let's say we want to make a function that adds together two variables and returns their sum. We could write:

```python
# This function adds two numbers and *returns* the result

def my_sum(a, b):
    c = a + b
    return c
    
```

When we call (use) the function, we provide values for the inputs:

```python
sum1 = my_sum(a=5, b=6)

print(sum1)
```

Notice that we have defined a variable, `sum1`, and set it equal to the function

When we do this, the variable will be assigned the value **returned** by the function -- i.e. the value of `c`

## Global and local variables

What happens if we try to print `c` directly (rather than printing the result in `sum1`)?

```python
# print(c)
```

We get an error: `c is not defined` 

Is this surprising? After all, we defined a variable called `c` in our function!

This error highlights the important difference between **global** and **local** variables

 - **global** variable: A variable defined in the main body of your code. Any code executed after the variable has been defined is able to 'see' the variable
 - **local** variable: A variable defined inside a function or other object. Its value is only accessible *within* the function or object

Since `a`, `b` and `c` have only been defined inside the function, these are local variables. We are unable to access them outside of the function

`sum1`, in contrast, is a global variable (defined outside the function), so we can print its value anywhere we like

Functions can see their own local variables **and** any global variables, as long as the global variable is already defined:

```python
def add_to_d(a, b):
    c = a + b + d
    return c

d = 2

print(d)

# d has been defined before function is called

sum2 = add_to_d(a=1, b=2)

print(sum2)
```

You can see in the above that the function is able to use the value of `d`, since it has already been defined globally.

This can be useful sometimes, but be careful! If `d` is redefined somewhere in your main code, the value used by your function may not be what you think it is!

## Required and optional keyword arguments

So far, the functions we have defined will only work if all keyword arguments are provided.

To make an input optional, we simply give it a default value:

```python
# Function with required parameter a and optional parameter b

def my_sum(a, b=0):
    c = a + b
    return c
    
```

```python
sum1 = my_sum(a=10)

print(sum1)


sum2 = my_sum(a=10, b=1)

print(sum2)
```

In this example, if we don't provide a value for `b`, the function will use the specified default value (in this case 0)

If we do provide a value, this will be used instead of the default

## Functions with different variable types

Keyword arguments do not have to be numeric types. They can be strings, lists, Boolean (i.e. True or False), etc -- or even another function!

But of course, the type of argument supplied must be compatible with the purpose of your function!

Here is an example function with different types of inputs:

```python
# A function to print the elements in a list one at a time, in forward or reverse order

# Takes two keyword arguments: a list (required) and whether to reverse it (optional)

def print_list(my_list, reverse=False):
    
    # if user wants to reverse list, use list.reverse() method, otherwise do nothing
    if reverse == True:
        my_list.reverse()
    else:
        pass
    
    # iterate over list and print each element
    for i in my_list:
        print(i)
        
    return
```

```python
test_list = [1,2,3,4,5,6,7,8,9]

print_list(test_list)
```

## Functions with an arbitrary number of variables

In the functions above, we specified the number of arguments in advance

But we can also write functions that can take an arbitrary number of variables

This could be useful if you do not know in advance how much data you need to work with!

We can do this with the following syntax using an asterisk `*`:

```python
# Function taking any number of arguments. Use an asterisk in defining the function:

def arbitrary_sum(*numbers):
    sum = 0
    for i in numbers:
        sum += i
    return sum
```

You can change the inputs to the function below to see that it works for any number of arguments:

```python
print(arbitrary_sum(1,2,4))
```

## The return statement in detail

All functions in Python return something. If you do not specify a value (or leave out the `return` statement entirely), your function will return a `None` value by default. Otherwise, it will return the value we specify.

We can also return more than one value from a function, and return different types:

```python
def sum_and_product(a, b):
    a_plus_b = a + b
    a_times_b = a * b
    return a_plus_b, a_times_b
```

The function above returns two values. What does this output look like?

It turns out we have a choice. When we call the function, we can either:

  - Provide the same number of variables as the number of values returned. Each returned value then goes to a separate variable
  
  - Provide a single variable. This will then contain a list of the values returned by the function.
  
You can see these two approaches below. Both are perfectly valid, and often a matter of taste!

```python
sum1, product1 = sum_and_product(5,5)

print(sum1)
print(product1)

all_in_one = sum_and_product(5,5)

print(all_in_one)
```

---

Return statements end the function call and return to the main code. Therefore any code in your function **after** the `return` will not be executed.

This can be convenient if you want to put conditions for what to return:

```python
def divide(a, b):
    
    # Avoid divide by zero error:
    if b != 0:
        return a/b
    
    return 0
    
```

```python
print(divide(a=10, b=2))
```

Change the values of `a` and `b` above -- the function will exit at the first `return` it encounters. Therefore, the code only reaches the second `return` if `b == 0`.

In this case, we could have achieved the same result using an `if ... else` block, but the properties of `return` mean we don't need the `else`.

Notice also: we didn't specify a local variable `c` this time: the `return` can take an expression directly and return its value if we prefer.

## Lambda functions

Python also provides a quick way to make short functions that can be defined in one line.

These are called lambda functions.

They can take any number of arguments, but can only have one expression.

They do not include a return statement; they always return the result of the expression.

For example:

```python
# Define a lambda function to double a number
# Arguments go before the colon, the expression goes after

doubler = lambda x: x*2

print(doubler(5))
```

```python
# Define a lambda function to add two numbers

adder = lambda a,b: a+b

print(adder(1,2))
```

You will notice that we could have equally defined these functions using the normal `def` syntax:

```python
def doubler(x):
    return x*2

def adder(a,b):
    return a+b
```

However, lambda functions can be very convenient if we need to create a function for temporary use, e.g. within another function.

We will see examples later when working with Pandas dataframes.
