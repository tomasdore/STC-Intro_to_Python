### **Advantages** of computer programming

- Computers do not get bored: automate repetitive tasks
  - e.g. "rename 100,000 files"
- Computers do not get tired
  - avoid human error, perform tasks overnight, etc
- Computers are calculators
  - solve difficult numerical problems that would take a human too long
- Computer code is reproducible
  - you may not remember how you solved a problem 6 months ago, but if you have the code you just run again

### What computers **cannot** do:

- Computers are not creative
  - at the highest level, problem-solving is a creative exercise (even mathematical problems)
  - they may never fully replace the talented human problem-solvers (not to mention music, art, literature...)
- Computers are not ethical
  - they simply execute the tasks that are given, so programmers have to be careful!
- Computers only know what you tell them
  - the results of any program are only as good as the input data
  - this can reinforce societal biases if e.g. machine learning training sets contain implicit bias


**So the best programmers still have to think critically, creatively and ethically about how to use code!**

## There are many programming languages out there: why Python?

**This is just for reference; you do not need to recognise all of these terms**

- High-level language
  - i.e. **syntax looks closer to natural language** rather than underlying machine code
- Emphasises **readability**, making use of indentation and white space
- Dynamically typed
  - don't need to decide at the beginning whether a certain variable will be an integer, decimal, alphabetic, etc; can change types on the fly
- Interpreted language
  - do not need to recompile every time you edit something
- Assigns memory automatically
  - saves work for developer
- Supports multiple approaches to programming
  - Procedural programming: control the flow of execution using loops and logic gates
  - Object-oriented programming: define methods attached to various data objects
- **Extensive functionality**
  - libraries exist for numerical and scientific tasks, image processing, machine learning and much more
- Portable
  - **code runs on Mac, Windows, Linux etc**
- Open source
  - **free to use** without an expensive license
- **Very popular, especially for data science and machine learning**
  - so lots of good resources online!

There are some disadvantages, e.g.

- Slower than compiled languages
- Can be memory-intensive

## Squashing bugs

**Computer code rarely works on the first try**

Finding the source of errors in your code is called 'de-bugging'. The name refers to finding actual dead bugs inside early computers!

You will spend a lot of your time as a programmer finding and fixing bugs, in your own code or in others.

<!-- ![debugging](https://imgs.xkcd.com/comics/fixing_problems.png) -->

![debugging](images/debugging.png) 

### Top tips for de-bugging

- Always read error messages carefully
  - The last line usually tells you the type of error
  - The first line usually tells you which line of your code caused the error
- Comment your code thoroughly so that it's clear what each part is supposed to do
- Tell your code to print outputs from intermediate steps (not just the final result)
  - You can check that values are what they should be
  - You will see how to print information below!
- The internet is your best friend
  - Look at the documentation pages for the Python packages you are trying to use -- they are full of examples!
  - Paste error messages into Google
  - Use the website Stack Overflow -- most common issues have been solved on there!


To the left of the code cells, you will notice some text like

`In [1]:`


This tells you in what order cells have been executed (highest number is most recent execution). Once a cell is run, your computer will 'remember' the results until you tell it otherwise. This can affect any cell that is executed afterwards, even if it appears earlier in the notebook.

Sometimes running cells out of sequence can have unintended consequences, so this number can be very useful to know what cells have already been run!

**To see this, run the cells above in a different order. You will see the output of the middle cell depends on the order of execution**

Jupyter notebooks are a very convenient way to run code a little bit at a time, but always be careful if you go back to earlier cells.
