## Variables

A **variable** has a fixed name, and a value that can change. We use the name to return the value.

For the mathematically-inclined, this is a lot like algebra: if I have an expression like y=x+1, the quantities named 'x' and 'y' can have many different values. In this case, the value of one variable (y) depends on the other (x).

A variable does not have to be a number, it just needs to have a name and a value.


---

### Some types of variables in Python:

- Numeric: integers (whole numbers), floating point or 'floats' (with decimal places), or complex numbers, e.g.
  - my_integer = 1
  - new_float = 3.14
- Text: called a string in programming. Always marked by quotation marks (can be single or double quotes), e.g.
  - string1 = 'word', 
  - string2 = "123", 
  - string3 = "This variable is a whole sentence"
    - NOTE: strings can contain numbers, but you can't use them for maths!
- Boolean: True or False
  - some_condition = True
- Sequences: lists or arrays of numbers/letters, and more complicated structures we will get to later!

---

In our next code, we will use a variable:

```python
# Define a variable called my name (in this case a string)
# Try changing it to your own name!
my_name = 'Matt'

# Print the user-defined string
print('Hello', my_name, '!')

# The print function will print everything inside the brackets, in order, but these must be separated by commas
```

A statement like `my_name = 'Matt'` is called **assignment** -- we assign a value to a variable.

### Best practice

Although you are free to choose variables, it will make your code **much** easier to read (and write) if you use meaningful names.

This is especially true when your code is very long: if you define a very generic variable name, by the end of the code it might not be easy to remember what it represents.

Let's say I have a variable that represents the number of movies starring Ryan Gosling.

An ok name would be something like `n_movies`.

A good name could be `n_movies_gosling`.

And you should avoid using an uninformative name like `n`.

## Changing variable type

You can change a variable from one type to another:


```python
# Define a string:
x = '33.3'

# Check the value and type of x
print(x)
print(type(x))
```

```python
# Change it to a float:
x = float(x)

# Check the value and type again
print(x)
print(type(x))

# it is now a number and can be used for maths:
print(x+1)
```

```python
# Change it to an integer - this will round the value to a whole number:
x = int(x)

# Check the value and type again
print(x)
print(type(x))
```

Note that not all conversions will work. E.g. you cannot change a string '33.3' directly to an integer, due to the decimal place:

```python
# # This cell will cause an error

# x = '33.3'

# x = int(x)
```

## Inputs

Very often, you will want the user to be able to provide a new value for some variable, without having to edit the code directly.

We do this using the `input` function:

```python
# The input function will prompt the user with whatever string is inside the brackets
my_name = input('Enter your name: ')

print('Hello,', my_name,'!')
```

### What about numbers?

```python
my_age = input('Enter your age: ')

# If you run the cells in order, the kernel will remember your name from the previous cell:
print(my_name, 'is', my_age, 'years old')
```


---

## The basic flow of a code

Most useful codes take the form:

    input -> task -> output
    
Here is a simple example:

```python
# We start with an input
my_age = input('Enter your age: ')

# Now we do a simple task: 

# Step 1. The input function always produces a string
# This next line changes my_age from a string (text) to an integer (a whole number)
my_age = int(my_age)

# Step 2. We now do some maths:
my_next_age = my_age + 1

# And produce an output:
print('At my next birthday, I will be',my_next_age)
```

---

## Expressions

A line like

    my_next_age = my_age + 1
    
is called an expression.

An expression is any group of variables (`my_age`) or constants (`1`) that together result in a value (`my_next_age`).

These are extremely common in programming, especially mathematical expressions like this one.

The basic maths expression usually involve:

    + (add)
    - (subtract)
    * (multiply)
    / (divide)
    % (remainder)
    ** (raise to the power of)

Here are some examples:

```python
x = 3
y = 2

a = x + y
b = x - y
c = x * y
d = x / y
e = x % y
f = x ** y

print('x + y =', a)
print('x - y =', b)
print('x * y =', c)
print('x / y =', d)
print('Remainder of x/y =', e)
print('x ^ y =', f)

# Note that division returns a float, rather than an integer
```

---

Some maths expressions also work on strings:

```python
x = 'super'
y = 'nova'

# This concatenates the strings together
z = x + y

print(z)
```

```python
w = 'repetition'

# This concatenates the string to itself 5 times
v = w * 5

print(v)
```
