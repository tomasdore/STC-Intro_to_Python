## Dictionaries

A very handy way to store and access data in Python is with dictionaries.

A dictionary is a set of keyword and value pairs. You use the keyword to access the value. Again the value can be of any Type, including another dictionary

The notation for a dictionary is a curly bracket:

```python
# Initialise a dictionary, with some keyword-value pairs

planet_masses = {'Earth': 1.0 , 'Jupiter': 317.8, 'Mercury': 0.0553}
# These are the masses of some planets, relative to the Earth

print(planet_masses)
```

We can access any value using the keyword, similar to how we used indices for lists. 

The advantage of a dictionary is that we don't need to care about *where* the value we want is, we just have to remember what we called it!

```python
# Remember that the name of the key is always a string, and needs quotation marks!
print(planet_masses['Jupiter'])
```

We can easily add new entries to a dictionary:

```python
planet_masses['Mars'] = 0.107

print(planet_masses)
```

Dictionaries are very useful for storing and retrieving data, but they do not permit e.g. easy math operations like lists

## Lists of dictionaries, dictionaries of lists

We have already encountered nested lists ('lists within lists'). But we can also nest mixed types of structured data.

A common structure in data science is a list in which each item is a dictionary:

```python
pavement_albums = [{'Title':'Slanted and Enchanted', 'Year':1992}, {'Title':'Crooked Rain, Crooked Rain', 'Year':1994}]
```

We can access each dictionary using the usual list indexing:

```python
print(pavement_albums[0])
```

Or to access only the title, we can combine numerical list indexing and dictionary keys:

```python
print(pavement_albums[0]['Title'])
```
