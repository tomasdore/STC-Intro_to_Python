# Programming Paradigms

A paradigm is like a philosophy informing how we write code. Usually, there are many different ways to solve a problem with code. Different paradigms help to shape which approach we choose to use.

The most common paradigms, both supported in Python, are

**Procedural programming**:

The code is organised as a sequence of instructions (do this, then do this) etc, with each block performing a prescribed task to solve the problem. 

**Object-oriented programming**:

Data are stored as 'objects' belonging to pre-defined 'classes'. These objects have a set of 'attributes' stored internally, which can be updated using built in 'methods'.



So far we have implicitly adopted a procedural approach. Here we will see the key characteristics of object-oriented programming.

---


# Classes and objects

The key to object oriented programming is to understand **classes**. These will be easiest to understand when we make our own example.

A class is like a template designed in advance to handle a particular data structure, with a set of properties called **attributes**.

You have seen examples of classes before! For example, the basic Python data types (e.g. string, list) are all classes.

A class can also have built-in functions. The syntax looks like `class_name.function_name()`.

For example, `list.reverse()` is a built-in function that operates on the `list` class.

You can see all the attributes and functions of a class using the `dir()` function:

```python
# See all the attributes (properties) and methods (functions) of the list class:

print(dir(list))
```


As well as methods (functions) that you might recognise (reverse, sort, etc) you can also see attributes of the object. This includes things like the class the object belongs too.

Attributes with a double underscore are internal to python and cannot be updated.

---

When we create a new list, e.g. `my_list = [x,y,z]`, the new list is an **object**. 

Any object is an **instance** (i.e. one specific example) of a class, created to represent a particular item of data.

The object **inherits** the properties of the class it represents.

Objects also inherit any built-in functions from their parent class. These are called **methods** of the object. 

The methods will update the internal state of the object -- e.g. reversing the list.

```python
# You can also use dir to look at the attribute of an object -- i.e. an instance of the class rather than the class itself:

my_list = [1,2,3]

print(dir(my_list))
```

```python
# Prove that the object is a member of the class 'list':

print(my_list.__class__)
```

## Creating a class

Object-oriented programming means creating your own classes and methods that you can use to represent and analyse your data.

Your class can have any attributes you want!

Just like loops, conditions and functions, we use indentation to define a class.

Consider the example below, where we create a class to hold data about an animal:

```python
# This first line defines the name of the class (syntax is similar to defining a function)

class Animal():

    # We give the class some basic attributes that any instance would have:
    
    name = ''
    species = ''
    colour = ''
    hunger = 0
        

    # We can define some functions that update the attributes of the class:
    
    def exercise(hunger):       

        # If the animal does some exercise, it gets hungry: the internal hunger attribute gets updated to 100%
        
        print('Fetch!')
        hunger = 100
        print('Hunger level: ', hunger)
        
        return hunger

    
    def feed(hunger):

        # If we feed the animal, it resets the hunger to zero.

        print('Tasty!')
        hunger = 0
        print('Hunger level: ', hunger)

        return hunger



# Confirm that our class exists:

print(Animal)
```

## Why define class-wide attributes? The `__init__` function

In the previous example, we had to set attribute values one-by-one after instantiating our object.

However, if you know that you will want to create many objects from the same class, with shared attributes, you can also assign values when creating each object.

This uses a special function called `__init__`:

```python
class Animal():

    # The class starts with a function called __init__ to initialise the attributes of the class
    # Attributes in the brackets must be specified when creating an object
    
    def __init__(self,name,species,colour):
    
        # 'self' is a special variable that will represent the object once it is instantiated
        # self.x means the attribute 'x' belonging to the object 'self'
    
        self.name = name
        self.species = species
        self.colour = colour
        self.hunger = 0
        

    # We can define some functions that update the attributes of the class
    # Notice that the function can take 'self' as input, and then access any attribute with self.x
    
    def exercise(self):       
        
        print('Fetch!')
        self.hunger = 100
        print('Hunger level: ', self.hunger)
        
        return self.hunger

    
    def feed(self):

        print('Tasty!')
        self.hunger = 0
        print('Hunger level: ', self.hunger)

        return self.hunger


    # If we use self as the input argument, we have to remember to use self.x throughout the function!
```

Now we can create objects much more quickly:

```python
# Create cat and dog objects in one line each:

ludo = Animal('Ludo', 'cat', 'orange')

fido = Animal('Fido', 'dog', 'black')
```

```python
# Check the objects have been correctly instantiated:

print(fido.species)
```

## Class inheritance

We can also define classes in a hierarchical manner. In the animal example, this could look something like:

 - 'Parent' class: Animal
   - 'Child' class: Cat, Dog, etc
     - object: your pet

A child class inherits the attributes of its parent class. But we can also add new methods to the child class.

Let's create a `Cat` class that inherits most of its methods and attributes from the `Animal` class:

```python
# To make this a child class, we just put the name of the parent class in the brackets:

class Cat(Animal):

    # Any attributes from the parent class that should be fixed for the child class go first:

    species = 'cat'

    
    # Then we use __init__ for the attributes we want to specify new values for:
    
    def __init__(self, name, colour):
        
        # Instead of redefining the attribute, we use `super()` to inherit from the parent class
        # Note we can use self.x to keep x fixed to the value provided at the beginning
        
        super().__init__(name, self.species, colour)

    
    # Finally we define any new methods or attributes specific to objects in the child class
    
    def meow(self):
        print('Meow!')
```

```python
# This time we instantiate our pet cat as an object of the Cat class

ludo = Cat('Ludo', 'orange')
```

Notice that the `hunger` variable is not mentioned in the `Cat` class. However, this will be inherited, with its default value, from the `Animal` parent class:

```python
print(ludo.hunger)
```

```python
ludo.exercise()
```

And now we can also use the new methods defined in the `Cat` class:

```python
ludo.meow()
```

## Modules: making your classes and functions available anywhere

If you create a very useful function or class, you may want to use it in many different codes.

You can avoid copying and pasting by making your code a Python **module**!

A module is a python file (extension `.py`) containing one or more classes or functions.

You can **import** these from the module easily!

Download the provided module, `animals.py`, in the same directory as this notebook.

Now we can load the `Dog` class from this module!

```python
from animals import Dog

fido = Dog('Fido', 'black')

fido.bark()
```

We can also define and import functions -- they don't need to be part of a class, making it very easy to reuse them:

```python
from animals import pet

pet('Ludo')
```
