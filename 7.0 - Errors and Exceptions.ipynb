As we have seen many times, coding results in errors. Usually these are unintended, but sometimes they are unavoidable or even desirable: for example, preventing undesirable outputs from an impossible calculation.

The goal when writing a program is to allow it to 'fail gracefully'. We can identify two cases:

- Errors: a fundamental issue that where Python cannot understand your code (syntax error)

- Exceptions: your code is written in valid Python syntax, but an operation cannot be completed successfully

**Syntax errors** will fail like this code below, and can only be fixed by spotting your programming errors and typos!

```python
# Example syntax error

print('I didnt check my brackets'))
```

## Exceptions: the try - except block

If you know places where your code may produce exceptions under some circumstances, we can think ahead and try to 'catch' it.


Python provides a convenient syntax for this:

    try:

        (your code)

    except:

        (what to do if it fails)

Note that the structure of this looks very similar to `if ... else`!

Consider a simple example:

```python
# Define a variable x
x = 1.4

# Convert it to an integer y
y = int(x)

# Print y
print('y = %d' %y)
```

```python
# Define a variable x
x = 'string'

# This fails if x is a string!
y = int(x)
```

Since we know that `y = int(x)` will give an error for some values of `x`, we can use `try ... except` to prevent our code from crashing and implement an emergency fallback option

```python
x = 1.4

try:
    y = int(x)
    print('y = %d' %y)
except:
    print('y is not a number!')
```

Try changing `x` to a string above, to see the different behaviour

---

## A note of caution!

The `try ... except` block will catch all exceptions -- even if the error is not what you think it is!

```python
# x is a valid number
x = 1.4

try:
    # int(x) is allowed
    y = int(x)
    
    # but opening a non-existent file is not!
    f = open('non-existent-file.txt', 'r')
    f.close()
    
    print('y = %d' %y)
    
except:
    print('y is not a number!')
```

The code above catches the exception. It prints the warning that the user has set ("y is not a number!"). But in this case, the error was caused by an invalid file operation.

Do not use `try ... except` blindly, but rather use it to try to catch **specific** errors.

Python includes several types of built-in exception types for different kinds of errors. 

For example, we can see from earlier that when we try to force `int('string')`, Python threw a ValueError.

You can see a full list of built-in exception types here: https://docs.python.org/3/library/exceptions.html

---

We can now use `try ... except` more carefully to respond differently to the expected possibility of a ValueError.

```python
x = 'string'

try:
    # int(x) is allowed
    y = int(x)
    
    # but opening a non-existent file is not!
    f = open('non-existent-file.txt', 'r')
    f.close()
    
    print('y = %d' %y)
    
# Specify what to do in case of a ValueError:
except ValueError:
    print('y is not a number!')
    
# Specify what to do in case of a generic error:
except Exception:
    print('Some other error occurred')
```

Change `x` above between a string type and a numeric type to see that different exceptions are caught!

---

## Try - Except - Else - Finally


We can extend the exception-handling block with additional steps to execute after the `try...except`.

    else: Do this if no exception
    
    finally: Always do this at the end
    
For example, change the type of `x` below to see the different behaviour of `else` and `finally`:

```python
x = 'string'

try:
    # int(x) is allowed
    y = int(x)
    print('y = %d' %y)
    
# Specify what to do in case of a ValueError:
except ValueError:
    print('y is not a number!')
    
# Specify what to do in case of a generic error:
except Exception:
    print('Some other error occurred')
    
else:
    print('Code executed with no exceptions')
    
finally:
    print('Done')
```

---

## Summary

- The `try` - `except` statements can catch most types of errors and rescue your code
- But be cautious that the errors may not be what you expect -- better to use specific exception types for complex code
